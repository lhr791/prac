## 合并(left_join)

```r
b<-c(1:10)
```

```r
c<-c(5:15)
```

```r
df1<-cbind(b,b^2)
```

```r
df2<-cbind(c,c^3)
```

```r
try(left_join(df1,df2)) #名字不一致，导致error
```

```r
colnames(df1)<-c("x1","x2")
```

```r
df1<-tbl_df(df1)
```

```r
colnames(df2)<-c("x1","x3")
```

```r
df2<-tbl_df(df2)
```

```r
left_join(df1,df2) #先匹配，后合并
```

```r
right_join(df1,df2)
```

```r
inner_join(df1,df2) #两个都有的，合并
```

```r
full_join(df1,df2)
```

# final

```r
df<-flights%>%mutate(speed=distance/air_time)%>%select(c("air_time","distance","carrier","speed"))%>%
group_by(carrier)%>%summarise_all(mean,na.rm=T)%>%arrange(desc(speed))
```

```r
ggplot(flights,aes(x=air_time,y=distance,color=carrier))+geom_point()
```

```r
ggplot(flights,aes(x=air_time,y=distance,color=carrier))+geom_jitter(alpha=0.3)
```

```r
ggplot(flights,aes(x=air_time,y=distance,color=carrier))+geom_point(shape=1)+facet_grid(.~carrier) #much better than the picture above
```

```r
ggplot(df,aes(x=speed))+geom_density()
```

```r
ggplot(df,aes(x=carrier,y=speed,fill=carrier))+geom_bar(stat="identity")
```

```r
ggplot(df,aes(x=carrier,y=speed,fill=speed))+geom_bar(stat="identity")+scale_fill_gradient(low="#ffe8e8",high="#ff9292")
```
